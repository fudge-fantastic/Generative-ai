```python
import os
import dotenv
import google.generativeai as genai
from IPython.display import display_markdown

dotenv.load_dotenv()  
api_key = os.environ.get("GOOGLE_API_KEY")
genai.configure(api_key=api_key)  

if not api_key:
    raise ValueError("API Key not found")
```

```python
for model in genai.list_models():
    if 'generateContent' in model.supported_generation_methods:
        print(model.name)
```

```python
def generate_text(input_text):
  dotenv.load_dotenv()  
  api_key = os.environ.get("GOOGLE_API_KEY")
  
  safety_settings = [
  {
    "category": "HARM_CATEGORY_HARASSMENT",
    "threshold": "BLOCK_NONE"
  },
  {
    "category": "HARM_CATEGORY_HATE_SPEECH",
    "threshold": "BLOCK_NONE"
  },
  {
    "category": "HARM_CATEGORY_SEXUALLY_EXPLICIT",
    "threshold": "BLOCK_NONE"
  },
  {
    "category": "HARM_CATEGORY_DANGEROUS_CONTENT",
    "threshold": "BLOCK_NONE"
  }
]

  genai.configure(api_key=api_key)  

  generation_config = {
      "temperature": 1,
      "top_p": 0.95,
      "top_k": 64,
      "max_output_tokens": 8192,
      "response_mime_type": "text/plain",
  }

  model = genai.GenerativeModel(
      model_name="gemini-1.5-pro-latest",
      generation_config=generation_config, safety_settings=safety_settings,
  )

  chat_session = model.start_chat(history=[])
  response = chat_session.send_message(input_text)
  display_markdown(f"{response.text}", raw=True) 
```

```python
generate_text("I need a solution for Trap the Rain Water - II")
```

```python
generate_text("I need a table that shows cute things, and write something appealing :)")
```

```python
generate_text("Write any complex code, where one can barely interpret!")
```
